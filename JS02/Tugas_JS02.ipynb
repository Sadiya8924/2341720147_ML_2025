```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/ml/dataset/wbc.csv'
```

```python
df = pd.read_csv(file_path)

df.head()
```

```python
print("Dimensi (baris, kolom):", df.shape)

df.info()

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
cols = df.columns.tolist()
print("Semua kolom:", cols)

n_rows = df.shape[0]
unique_count = df.nunique()
non_informative = unique_count[unique_count == n_rows].index.tolist()
print("\nKolom dengan semua nilai unik (potensi ID/non-informative):", non_informative)
```

```python
df_proc = df.copy()

cols_to_drop = non_informative.copy()

if 'Unnamed: 32' in df_proc.columns:
    cols_to_drop.append('Unnamed: 32')

cols_to_drop = list(dict.fromkeys(cols_to_drop))

print("Kolom yang akan dibuang:", cols_to_drop)

df_proc.drop(columns=cols_to_drop, inplace=True, errors='ignore')

print("\nKolom tersisa setelah drop:", df_proc.columns.tolist())
```

```python
print("Apakah 'diagnosis' ada di dataset?", 'diagnosis' in df_proc.columns)

num_cols = df_proc.select_dtypes(include=[np.number]).columns.tolist()
print("Kolom numerik terdeteksi:", num_cols)

cat_cols = df_proc.select_dtypes(include=['object', 'category']).columns.tolist()
print("Kolom kategorikal terdeteksi:", cat_cols)
```

```python
print("Apakah 'diagnosis' ada di dataset?", 'diagnosis' in df_proc.columns)

num_cols = df_proc.select_dtypes(include=[np.number]).columns.tolist()
print("Kolom numerik terdeteksi:", num_cols)

cat_cols = df_proc.select_dtypes(include=['object', 'category']).columns.tolist()
print("Kolom kategorikal terdeteksi:", cat_cols)
```

```python
print("Unique values in diagnosis:", df_proc['diagnosis'].unique())

df_proc['diagnosis_encoded'] = df_proc['diagnosis'].map({'M':1, 'B':0})

print(df_proc[['diagnosis','diagnosis_encoded']].head())
```

```python
num_cols_all = df_proc.select_dtypes(include=[np.number]).columns.tolist()

if 'diagnosis_encoded' in num_cols_all:
    features_to_scale = [c for c in num_cols_all if c != 'diagnosis_encoded']
else:
    features_to_scale = num_cols_all.copy()

print("Kolom yang akan distandarkan (features):", features_to_scale)

scaler = StandardScaler()

df_proc_scaled = df_proc.copy()
df_proc_scaled[features_to_scale] = scaler.fit_transform(df_proc_scaled[features_to_scale])

df_proc_scaled.head()
```

```python
print(df_proc_scaled[features_to_scale].describe().loc[['mean','std']])
```

```python
out_path = '/content/drive/MyDrive/ml/dataset/wbc_preprocessed.csv'
df_proc_scaled.to_csv(out_path, index=False)
print("File preprocessed tersimpan di:", out_path)
```
