##Memuat Data

```python
df_voice = pd.read_csv('/content/drive/MyDrive/ml/dataset/voice.csv')
df_voice.head()
```

##Persiapan Data (Encoding & Pemisahan Fitur)

```python
from sklearn.preprocessing import LabelEncoder

# Asumsi kolom target bernama 'label' (isinya 'male'/'female')
# Jika nama kolomnya beda, GANTI 'label' di bawah ini
target_column = 'label'

# Mengubah label 'male'/'female' menjadi 0/1
le = LabelEncoder()
df_voice[target_column] = le.fit_transform(df_voice[target_column])

# Pisahkan antara Fitur (X) dan Target (y)
X = df_voice.drop(target_column, axis=1) # Semua kolom KECUALI 'label'
y = df_voice[target_column]               # HANYA kolom 'label'

print("Fitur (X) berhasil dipisah:")
print(X.head())
print("\nTarget (y) berhasil dipisah:")
print(y.head())
```

##Persiapan Tabel Hasil

```python
# Kita siapkan list kosong untuk menampung hasil akurasi
hasil_akurasi = []
```

##Pemodelan Split 70:30

```python
print("--- Memulai Pemodelan Split 70:30 ---")
# 1. Split data 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print(f"Ukuran data latih: {X_train.shape[0]}, Ukuran data tes: {X_test.shape[0]}")

# 2. Model Kernel Linear
print("Melatih model Linear...")
model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
y_pred_linear = model_linear.predict(X_test)
acc_linear = accuracy_score(y_test, y_pred_linear)
hasil_akurasi.append({'Split': '70:30', 'Kernel': 'Linear', 'Akurasi': acc_linear})
print(f"Akurasi Linear (70:30): {acc_linear:.4f}")

# 3. Model Kernel Polynomial
print("Melatih model Polynomial...")
model_poly = SVC(kernel='poly')
model_poly.fit(X_train, y_train)
y_pred_poly = model_poly.predict(X_test)
acc_poly = accuracy_score(y_test, y_pred_poly)
hasil_akurasi.append({'Split': '70:30', 'Kernel': 'Polynomial', 'Akurasi': acc_poly})
print(f"Akurasi Polynomial (70:30): {acc_poly:.4f}")

# 4. Model Kernel RBF
print("Melatih model RBF...")
model_rbf = SVC(kernel='rbf')
model_rbf.fit(X_train, y_train)
y_pred_rbf = model_rbf.predict(X_test)
acc_rbf = accuracy_score(y_test, y_pred_rbf)
hasil_akurasi.append({'Split': '70:30', 'Kernel': 'RBF', 'Akurasi': acc_rbf})
print(f"Akurasi RBF (70:30): {acc_rbf:.4f}")

print("--- Selesai Split 70:30 ---")
```

##Pemodelan Split 80:20

```python
print("\n--- Memulai Pemodelan Split 80:20 ---")
# 1. Split data 80:20
X_train_80, X_test_80, y_train_80, y_test_80 = train_test_split(X, y, test_size=0.20, random_state=42)
print(f"Ukuran data latih: {X_train_80.shape[0]}, Ukuran data tes: {X_test_80.shape[0]}")

# 2. Model Kernel Linear
print("Melatih model Linear...")
model_linear_80 = SVC(kernel='linear')
model_linear_80.fit(X_train_80, y_train_80)
y_pred_linear_80 = model_linear_80.predict(X_test_80)
acc_linear_80 = accuracy_score(y_test_80, y_pred_linear_80)
hasil_akurasi.append({'Split': '80:20', 'Kernel': 'Linear', 'Akurasi': acc_linear_80})
print(f"Akurasi Linear (80:20): {acc_linear_80:.4f}")

# 3. Model Kernel Polynomial
print("Melatih model Polynomial...")
model_poly_80 = SVC(kernel='poly')
model_poly_80.fit(X_train_80, y_train_80)
y_pred_poly_80 = model_poly_80.predict(X_test_80)
acc_poly_80 = accuracy_score(y_test_80, y_pred_poly_80)
hasil_akurasi.append({'Split': '80:20', 'Kernel': 'Polynomial', 'Akurasi': acc_poly_80})
print(f"Akurasi Polynomial (80:20): {acc_poly_80:.4f}")

# 4. Model Kernel RBF
print("Melatih model RBF...")
model_rbf_80 = SVC(kernel='rbf')
model_rbf_80.fit(X_train_80, y_train_80)
y_pred_rbf_80 = model_rbf_80.predict(X_test_80)
acc_rbf_80 = accuracy_score(y_test_80, y_pred_rbf_80)
hasil_akurasi.append({'Split': '80:20', 'Kernel': 'RBF', 'Akurasi': acc_rbf_80})
print(f"Akurasi RBF (80:20): {acc_rbf_80:.4f}")

print("--- Selesai Split 80:20 ---")
```

##Split 80:20

##Model RBF

```python
# Membuat model RBF dengan parameter default (dasar)
model_rbf_p5_base = SVC(kernel='rbf')
model_rbf_p5_base.fit(X_train_p5, y_train_p5)

# Prediksi dan catat akurasi
y_pred_p5_base = model_rbf_p5_base.predict(X_test_p5)
acc_p5_base = accuracy_score(y_test_p5, y_pred_p5_base)

print(f"Akurasi Model RBF (Base/Default): {acc_p5_base:.4f}")
```

##Eksperimen Hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV

# Tentukan 'kandidat' parameter yang ingin kita coba
# C: Seberapa besar 'hukuman' untuk kesalahan klasifikasi
# gamma: Seberapa jauh pengaruh satu data latihan
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001]
}

print("Memulai Hyperparameter Tuning (GridSearch)... Ini mungkin butuh waktu...")
# GridSearchCV akan mencoba semua (4x4=16) kombinasi parameter
# cv=5 artinya menggunakan Cross-Validation 5-fold
grid_search = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=5, verbose=1)
grid_search.fit(X_train_p5, y_train_p5)

print("\nTuning Selesai.")
print("Parameter RBF terbaik yang ditemukan:")
print(grid_search.best_params_)
```

##Catat Performa Akurasi Model Hasil Tunning

```python
# Ambil model terbaik dari hasil GridSearch
model_terbaik_p5 = grid_search.best_estimator_

# Uji model terbaik ke data tes
y_pred_p5_tuned = model_terbaik_p5.predict(X_test_p5)
acc_p5_tuned = accuracy_score(y_test_p5, y_pred_p5_tuned)

print("\n--- HASIL AKURASI TUGAS 2 (SIANG/MALAM) ---")
print(f"Akurasi RBF (Default): {acc_p5_base:.4f}")
print(f"Akurasi RBF (Hasil Tuning): {acc_p5_tuned:.4f}")
```
